Elimina identificadores (id) porque no aportan valor al modelo.

Convierte variables categóricas a dummies → por ejemplo sexo se transforma en sexo_M (0 o 1).

Uso drop_first=True para evitar la trampa de la multicolinealidad (queda una categoría menos).

Deja todas las variables numéricas listas (edad, salario_mes, anos_experiencia, etc.).

```python
import pandas as pd

# Lista de columnas que compartiste
columnas = ['edad', 'abandono', 'viajes', 'departamento', 'distancia_casa', 'educacion',
            'carrera', 'empleados', 'id', 'satisfaccion_entorno', 'sexo', 'implicacion',
            'nivel_laboral', 'puesto', 'satisfaccion_trabajo', 'estado_civil', 'salario_mes',
            'num_empresas_anteriores', 'mayor_edad', 'horas_extra', 'incremento_salario_porc',
            'evaluacion', 'satisfaccion_companeros', 'horas_quincena', 'nivel_acciones',
            'anos_experiencia', 'num_formaciones_ult_ano', 'conciliacion', 'anos_compania',
            'anos_en_puesto', 'anos_desde_ult_promocion', 'anos_con_manager_actual']

# Clasificación
identificadores = ['id']
categ = ['abandono', 'viajes', 'departamento', 'sexo', 'puesto', 'estado_civil', 'horas_extra']
numericas = [col for col in columnas if col not in categ + identificadores]

print("🔹 Identificadores:", identificadores)
print("🔹 Categóricas:", categ)
print("🔹 Numéricas:", numericas)

# --- Limpieza del DataFrame ---
# Eliminamos identificadores porque no sirven para análisis predictivo
df = df.drop(columns=identificadores, errors='ignore')

# Convertimos variables categóricas en dummies (One-Hot Encoding)
df_ml = pd.get_dummies(df, columns=categ, drop_first=True)

# Mostramos resultado
print("✅ DataFrame transformado listo para ML:")
print(df_ml.head())

```
